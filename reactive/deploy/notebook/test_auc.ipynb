```python
import pandas as pd
```

```python
from sklearn import metrics
```

```python
label = pd.read_parquet('../data/01-06-2023/REACTIVATED')
pred = pd.read_parquet('../out/SCORE_01-06-2023')
```

```python
label = label.rename(columns={'FTR_VAL': 'LABEL'})
label['LABEL'] = label['LABEL'].astype(int)
```

```python
df = pred.merge(label, how='inner', left_on='CUSTOMER_CDE', right_on='CUSTOMER_CDE')
```

```python
y_true = df['LABEL'].values
y_pred = df['SCORE'].values
```

```python
metrics.
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y_true, y_pred)
metrics.auc(fpr, tpr)
```

```python
display = metrics.RocCurveDisplay.from_predictions(y_true, y_pred, plot_chance_level=True)
```
